```python
print("-------------------------This is a contact book-------------------------")

class ContactBook:
    def __init__(self):
        self.contacts = {}

    def addContact(self,name,phone,email,address):
      self.contacts[name]={'Phone':phone,'Email':email,'Address':address}

    def viewContact(self):
      for name, description in self.contacts.items():
            print(f"{name}: {description['Phone']} | {description['Email']} | {description['Address']}")

    def search(self,key):
      for name, description in self.contacts.items():
        if key in(name,description['Phone']):
          print(f"{name}: {description['Phone']} | {description['Email']} | {description['Address']}")
          return

        else:
          print("Record not found")

    def updateContact(self,name,detail,updated):
      if name in self.contacts:
        self.contacts[name][detail]=updated
        print(f"{name}'s {detail} changed to {updated}")
      else:
        print("Record not found")

    def deleteContact(self, name):
      if name in self.contacts:
        del self.contacts[name]
        print(f"{name} deleted from contacts.")
      else:
        print("Contact not found.")

a=ContactBook()
while(1):
  print("1. Add Contact")
  print("2. View Contact")
  print("3. Search Contact")
  print("4. Update Contact")
  print("5. Delete Contact")


  print()

  b=int(input("Choose from the above mentioned options"))
  if(b==1):
    name=input("Enter the name of the person")
    phone=input("Enter the phone number of the person")
    email=input("Enter the email of the person")
    address=input("Enter the address of the person")
    a.addContact(name,phone,email,address)

  elif(b==2):
    print("Here is the list of all the contacts in the contact book")
    print()
    a.viewContact()

  elif(b==3):
    key=input("Enter the name or phone number of the person that you want to search")
    a.search(key)

  elif(b==4):
    name=input("Enter the name of the person for record updation")
    detail=input("What attribute of the person do you want to change")
    new_value=input("Enter the updated value")
    a.updateContact(name,detail,new_value)

  elif(b==5):
    name=input("Enter the name of the person that you want to delete")
    a.deleteContact(name)

  else:
    print("Invalid option selected")

  c=input("Do you want to continue? (Y/y or N/n)")
  if(c=="Y" or c=="y"):

    print("OK")

  elif(c=="N" or c=="n"):
    break
```
